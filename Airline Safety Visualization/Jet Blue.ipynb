```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'http://www.airsafe.com/airline.htm'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text)

table = soup.find_all('table')

rows = table[1].find_all('tr')

th = rows[0].find_all('th')
columns = [i.text for i in th]
columns
data = []
for tr in rows[1:]:
    td = tr.find_all('td')
    row = [i.text for i in td]
    data.append(row)
data
```

```python
event_df = pd.DataFrame(data, columns= columns)
event_df.to_csv('events.csv', index= False)
```

```python
url = 'http://web.mit.edu/airlinedata/www/2020%2012%20Month%20Documents/Traffic%20and%20Capacity/System%20Total/Total%20System%20Revenue%20Passenger%20Miles.htm'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text)

table = soup.find('table')

rows = table.find_all('tr')
rows
# th = rows[0].find_all('th')
# columns = [i.text for i in th]
# columns
data = []
for tr in rows[3:]:
#     td = tr.find_all('td')
    row = [i.text for i in tr]
    data.append(row)
data
```

```python
rpm_df = pd.DataFrame(data).replace(['\n', ''], pd.NA).dropna(axis= 1, how= 'all').dropna(how= 'all').replace(pd.NA, '')
rpm_df = rpm_df.applymap(lambda x : x.strip().replace(',','').replace('-',''))
rpm_df
```

```python
rpm_df.iloc[0,0]= 'year'
rpm_df = rpm_df.transpose()
rpm_df
```

```python
rpm_df.columns = rpm_df.iloc[0]
```

```python
rpm_df = rpm_df.drop(1, axis= 0)
rpm_df
```

```python
rpm_df.to_csv('rpm.csv', index= False)
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_JetBlue_destinations')
response
```

```python
dest_df = pd.DataFrame(data)
dest_df.iloc[:,:2] = dest_df.iloc[:,:2].fillna(method='ffill',axis= 1)
```
