```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://aviation-safety.net/statistics/period/stats.php?cat=A1'
response = requests.get(url)
response
```

```python
with open('Aviation Safety Network _ Statistics _ By period.html') as f:
    dump = f.read()
```

```python
soup = BeautifulSoup(dump)

table = soup.find('table')

rows = table.find_all('tr')
th = rows[0].find_all('th')
columns = [i.text for i in th]

data = []
for tr in rows[1:]:
#     td = tr.find_all('td')
    row = [i.text for i in tr]
    data.append(row)
```

```python
crash = pd.DataFrame(data, columns=columns)
crash
```

```python
crash.to_csv('crash.csv', index=False)
```
